```python
import numpy as np
```

```python
array = np.random.randint(0, 100000, 1000)
array.sort()

target_val = array[np.random.randint(0, 1000, 1)]
```

```python
def search(val, array, begin, end, ratio):
    """ returns location and cumulative cost """

    n = end - begin

    if n == 1:
        return begin, begin
        
    guess = int(np.trunc(begin + ratio * n))
    
    if val == array[guess]:
        return guess, guess
    elif val < array[guess]:
        location, cost = search(val, array, begin, guess - 1, ratio)
        return location, cost + guess
    elif val > array[guess]:
        location, cost = search(val, array, guess + 1, end, ratio)
        return location, cost + guess
        
```

```python
x = []
y = []
for r in np.linspace(0.1, 0.9, 20):
    results = np.array([search(array[np.random.randint(0, 1000, 1)], array, 0, 1000, r) for i in range(1000)])
    x.append(r)
    y.append(results[:,1].mean())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10, 10))
plt.plot(x, y)
```
